```python
# ======================================================
#     VELKÝ PROGRAM PRACUJÍCÍ S JEDNÍM SEZNAMEM
# ======================================================

print("\n=== Vytvářím hlavní seznam AI nástrojů ===\n")

# 1) Vytvoření jednoho velkého seznamu
seznam = ["ChatGPT", "Claude", "Gemini", "Llama", "Mistral", "Perplexity", "AlphaCode", "Copilot", "Reka", "DeepSeek"]

print("Původní seznam:", seznam)
print("-" * 60)


# 2) Přidání několika položek
print("\n=== Přidávám nové položky ===")
seznam.append("Command R")
seznam.append("WizardLM")
seznam.append("OpenAI o1")
print("Seznam po přidání:", seznam)
print("-" * 60)


# 3) Odstranění prvního prvku
print("\n=== Odebírám první prvek ===")
odstraneny = seznam.pop(0)
print(f"Odstraněný prvek: {odstraneny}")
print("Seznam po odebrání:", seznam)
print("-" * 60)


# 4) Seřazení listu
print("\n=== Seřazuji seznam abecedně ===")
seznam.sort()
print("Seřazený seznam:", seznam)
print("-" * 60)


# 5) Vypsání délky seznamu
print("\n=== Délka seznamu ===")
print("Počet prvků:", len(seznam))
print("-" * 60)


# 6) Postupný výpis seznamu s čísly
print("\n=== Výpis všech položek ===")
for i, prvek in enumerate(seznam, start=1):
    print(f"{i}. {prvek}")

print("\n=== Program dokončen ===\n")
```
